The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search 
for guitars.
2. In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
from selenium.common.exceptions import ElementNotInteractableException,NoSuchElementException
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
urls=[]
x=range(1)
for i in x:
    print('scraping page',i+1)
    url=driver.find_elements_by_xpath('//h2[@class="title title-6"]//a')
    for i in url:
        urls.append(i.get_attribute('href'))    
    next_btn=driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[4]/div/div/div[33]/div[5]')
    next_btn.click()
```

```python
len(urls)
```

```python
#name of the hostel
name_hostel=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        name=driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/section/div[4]/div/div[1]/h1/div')
        name_hostel.append(name.text)
    except NoSuchElementException:
        name_hostel.append('-')
#done
```
