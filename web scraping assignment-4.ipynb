```python
#connect to web driver
driver=webdriver.Chrome(r'C:\Users\Aditya\OneDrive\Desktop\chromedriver.exe')
time.sleep(4)

url="https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(6)
driver.maximize_window()
```

```python
view_all_data=driver.find_element_by_xpath('/html/body/table[1]/tbody/tr/td[2]/span[2]/a')
view_all_data.click()
```

```python
#name
names=[]
name=driver.find_elements_by_xpath('//p[@class="normal"]/b/a[1]')
for i in name:
    if i.text is None:
        names.append('-')
    else:
        names.append(i.text)

#type of data        
Types=[]
types=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p")
for i in types:
    if i.text is None:
        Types.append('-')
    else:
        Types.append(i.text)
        
#task    
Task=[]
task=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p")
for i in task:
    if i.text is None:
        Task.append('-')
    else:
        Task.append(i.text)
        
#attribute type        
attribute=[]
attr=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p")
for i in attr:
    if i.text is None:
        attribute.append('-')
    else:
        attribute.append(i.text)
        
#no of instances        
instances=[]
insts=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p")
for i in insts:
    if i.text is None:
        instances.append('-')
    else:
        instances.append(i.text)
        
#no of attribute
no_of_attributes=[]
n_attr=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p")
for i in n_attr:
    if i.text is None:
        no_of_attributes.append('-')
    else:
        no_of_attributes.append(i.text)
        
        
        
#year       
year=[]
yr=driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p")
for i in yr:
    if i.text is None:
        year.append('-')
    else:
        year.append(i.text)
        
```

```python
len(Types)
```

```python
#removing heading 
del Types[0]

del Task[0]

del attribute[0]

del instances[0]

del no_of_attributes[0]

del year[0]
```

```python
data_sets=pd.DataFrame({'Dataset name':names,'Data type':Types,'Task':Task,'Attribute type':attribute,' No of instances':instances,'No of attribute':no_of_attributes,'year':year})
data_sets
```
